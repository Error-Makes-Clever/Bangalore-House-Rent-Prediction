```python
import pandas as pd
import numpy as np
```

```python
dfn = pd.DataFrame()
dfs = pd.DataFrame()
dfe = pd.DataFrame()
dfw = pd.DataFrame()
dfc = pd.DataFrame()

for i in range(5,101):
    d1 = pd.read_excel(f'99acres_Bangalore_North_scraped_data/99acres_rentals_page_{i}.xlsx')
    d1.insert(1, 'Region', 'Bangalore North')
    d2 = pd.read_excel(f'99acres_Bangalore_South_scraped_data/99acres_rentals_page_{i}.xlsx')
    d2.insert(1, 'Region', 'Bangalore South')
    d3 = pd.read_excel(f'99acres_Bangalore_East_scraped_data/99acres_rentals_page_{i}.xlsx')
    d3.insert(1, 'Region', 'Bangalore East')
    dfn = pd.concat([dfn,d1] , ignore_index=True)
    dfs = pd.concat([dfs,d2] , ignore_index=True)
    dfe = pd.concat([dfe,d3] , ignore_index=True)

for i in range(5,43):
    d1 = pd.read_excel(f'99acres_Bangalore_West_scraped_data/99acres_rentals_page_{i}.xlsx')
    d1.insert(1, 'Region', 'Bangalore West')
    dfw = pd.concat([dfw,d1] , ignore_index=True)

for i in range(5,22):
    d1 = pd.read_excel(f'99acres_Bangalore_Central_scraped_data/99acres_rentals_page_{i}.xlsx')
    d1.insert(1, 'Region', 'Bangalore Central')
    dfc = pd.concat([dfc,d1] , ignore_index=True)
```

```python
df = pd.concat([dfe,dfs,dfn,dfw,dfc],ignore_index=True)
```

```python
df = df.iloc[np.array(~pd.DataFrame(df['Link'].str.split('-').str.get(-1)).duplicated()) , :].reset_index(drop = True)
```

```python
df.drop(columns=['Location'] , inplace=True)
```

```python
df['Link'] = pd.DataFrame(df['Link'].str.split('-').str.get(-1))
```

```python
df.rename(columns= {'Link':'Property_ID'} , inplace=True)
```
